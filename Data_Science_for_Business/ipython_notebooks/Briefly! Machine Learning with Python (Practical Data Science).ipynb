Briefly: An Intro to Machine Learning (with Python)
===================================================

Through these notes, we'll review some of the basic concepts of machine learning (ML), show how techniques from machine learning can be called from python using the superb [scikit-learn](http://scikit-learn.org/stable/) library.

Machine learning has enjoyed increasing success and popularity in recent years, and constitutes one of the most crucial tools of the data scientist. Before further discussion on the details and applications of machine learning, it is useful to define it more formally. ML is a subfield of [artificial intelligence](http://en.wikipedia.org/wiki/Artificial_intelligence) noted for it's reliance on tools from statistics. In a nutshell, machine learning is concerned with how computer programs can learn from observations- data- find patterns within these observations, and generalize these patterns to new, previously unseen data. A more concise definition of ML from noted machine learning professor Tom Mitchell: "A computer program is said to learn from experience E with respect to some class of tasks T and performance measure P, if its performance at tasks in T, as measured by P, improves with experience E".

Machine Learning is a field with a history going back more than 60 years, to the early days of computer science with the perceptron algorithm. Today there are many subfields of machine learning considering different data structures, situations, relationships amongst data. Because the field of ML has become so broad, in these notes we must limit of coverage to a few of the most common, useful ML tools: classification, regression, and clustering. Before going into detail, let's go through some motiviating examples of each of these. 

For these examples, and throughout this document, we'll be using some [pre-configured datasets and dataset generators](http://scikit-learn.org/stable/datasets/) that ship with scikit learn. These datasets lower the cost of entry when trying machine learning- one of the most challenging- and crucial- aspects of building a ML system is loading raw data and transforming it into a format that can be consumed by ML models in order to provide accurate generalizations. More on these pre-rolled datasets and data ingestion generally below. 

Detailed documentation of all scikit-learn classes and functions can be [found here](http://scikit-learn.org/0.12/modules/classes.html).

Classification
--------------

Classification (often called categorization) is the most common task encountered in machine learning. Classification is concerned with assigning data to one of several pre-defined categories (or classes). When performing classification, there is a training phase in which pre-categorized data- usually called labeled data is provided to a model building system. The training system looks at this data and attempts to build a model that will accurately predict class labels on future observations that have missing labels. 

Below we see a toy problem in which we generate artificial two "blobs" of two dimensional data, then fit a line that can separate the two blobs. This line is a model that can be used to generalize, making future predictions. Given a new point from the same underlaying problem, this line could be used to assign a predicted membership into one  of the two blobs, depending on which side of the line on which the point falls.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
from sklearn.datasets.samples_generator import make_blobs

# we create 50 separable points
X, Y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)

# fit the model
# different types of models have different options that can be specified in their constructor
model = SGDClassifier(loss="hinge", alpha=0.01,
                      n_iter=200, fit_intercept=True)

# take the model and fit it to the data
model.fit(X, Y)

# plot the line, the points, and the nearest vectors to the plane
xx = np.linspace(-1, 5, 10)
yy = np.linspace(-1, 5, 10)

X1, X2 = np.meshgrid(xx, yy)
Z = np.empty(X1.shape)
for (i, j), val in np.ndenumerate(X1):
    x1 = val
    x2 = X2[i, j]
    p = model.decision_function([x1, x2])
    Z[i, j] = p[0]
levels = [-1.0, 0.0, 1.0]
linestyles = ['dashed', 'solid', 'dashed']
colors = 'k'

ax = plt.axes()
ax.contour(X1, X2, Z, levels, colors=colors, linestyles=linestyles)
ax.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired) # training data colored according to the label

ax.axis('tight')
plt.show()
```

Regression
----------

Regression is similar to classification in the sense that a model maps the obvervable features of data to a label value to be predicted some time in the future. However, while in classification the label is one of several discrete values, in regression the label is a continuous real-valued number. Below is an example of training a model on artificial training data. As before, a line is fit to the observations, however, in this case, the model is used to predict the y-value associated with a given x-value. 

```python
from sklearn.linear_model import LinearRegression

a = 0.5
b = 1.0

# x from 0 to 30
x = 30 * np.random.random(20)

# y = a*x + b with noise
y = a * x + b + np.random.normal(size=x.shape)

# create a linear regression model
model = LinearRegression()
model.fit(x[:, None], y)

# predict y from the data
x_new = np.linspace(0, 30, 100)
y_new = model.predict(x_new[:, None])

# plot the results
ax = plt.axes()
ax.scatter(x, y)
ax.plot(x_new, y_new)

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.axis('tight')
plt.show()
```

Because classification and regression are both sub-fields of machine learning where the ability from a model to generalize is acheived by building an understanding of examples with their associated true labels, they are often bundled together, along with several other more esoteric ML sub-fields, into a topic called "Supervised Learning".

Clustering
----------

In contrast to classification and regression, supervised learning, [clustering](http://en.wikipedia.org/wiki/Cluster_analysis) doesn't incorporate any label knowledge or ground truth information when building models. Rather, clustering is concerned with finding innate structure within data, finding elements of data- clusters- that have a high degree of similarity, and grouping them together. Below is a clustering of three random "blobs" of two-dimensional data. We see that the algorithm is able to identify a reasonably good grouping of data. Through this grouping comes a better understanding of a problem space, and potentially applications involving similar data points. 

```python
from sklearn.cluster import KMeans

centers = [[1, 1], [-1, -1], [1, -1]]
n_clusters = len(centers)
X, labels_true = make_blobs(n_samples=3000, centers=centers, cluster_std=0.7)

k_means = KMeans(init='k-means++', n_clusters=3, n_init=10)
k_means.fit(X) # the the model to the data

k_means_labels = k_means.labels_
k_means_cluster_centers = k_means.cluster_centers_
k_means_labels_unique = np.unique(k_means_labels)

ax = plt.axes()
colors = ['#4EACC5', '#FF9C34', '#4E9A06']
for k, col in zip(range(n_clusters), colors):
    my_members = k_means_labels == k
    cluster_center = k_means_cluster_centers[k]
    ax.plot(X[my_members, 0], X[my_members, 1], 'w',
            markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=6)
plt.show()
```

We've now seen a few toy examples of machine learning in python. Let's try something a bit more meaty- trying to decypher human handwriting given a digital image of that writing. In Practice: Using `scikit.learn`
---------------------------------

So far, we've seen a lot of code for building and using predictive models with scikit.learn. This may appear a bit daunting. Fortunately, the steps for training models are pretty consistent. While the various predictive models will differ in details, using different internal mechanisms to represent the problem being considered, the developer or data scientist can use them in a more or less uniform way. 

The first step in building a data mining model in python is chosing the actual model type. Here python has [a lot of options](http://scikit-learn.org/0.13/supervised_learning.html#supervised-learning). The documentation can help you choose the right model for your task, but generally there is going to be some trial and error.

Once you have chosen a model to use, you create an example of the model, by calling a special function called it's "constructor". In the above digit prediction task, we called the constructor for the `SVC` classifier in this line: 

```python
# a "support vector machine" predictive model
model = SVC(gamma=0.001, C = 100.0)
```

Note that we're passing two arguments to this function, `gamma` and `C`. These are optional values called "hyper-parameters" that define how the details of a model behave. The documentation for a particular model type will tell you what you need to know about the various hyper-parameters that it has available. Note that because these hyper-parameters are optional, there are default values that are often decent first things to test.

Once you have an instance of a model, ready to be trained, the next thing we need to do is train it! In the case of supervised learning problems, all we have to do as pass data that represents the examples to be modeled and their targets (labels) to the model's `fit` function:

```python
model.fit(digits.data, digits.target)
```

In the above example, `digits.data` is an list of lists, where each of the inner lists contains the feature values representing a single example. Because, deep in the internals of a machine learning algorithm is some math, ready to crunch on numbers, these input lists must contain all numeric types. Additionally, each list must contain the same number of "features".

`digits.target` represents the label values for the traning data passed to the model's `fit` method. There should be one entry in this list per example, there are as many labels in the target list as there are feature lists in the input data. Target values are numeric. In the case of regression problems, the target can be any floating point number. In the case of classificaiton, the target must be an integer. 

Note that in practice, much of the time a data scientist spends on a problem will be transforming a real world problem into the mathematical representation required by machine learning algorithms. Fortunately, scikit.learn helps you out a ton, more on this soon. 

Once you have a trained model, you'll usually want to use this model to make predictions. You can gather labels for a list of unlabeled training examples using the `predict` method:

```python
model.predict(some_new_data)
```

In the above, `some_new_data` is a list of lists, representing the features for new instances to be labeled. Because these will be used as input to the same mathematical functions that were constructed using your training data and the model's `fit` method, each of these feature lists should be the same size as the training data, and, more fundamentally, should be the output of the same feature generation process. 

In addition to just getting the most likely label for some data, for classification probelms, we often want to get class membership probabilities. We can do this with the `predict_proba` and `predict_log_proba` methods. These are used in much the same way as the `predict` method, only the output will be class membership probabilities or log-probabilities, respectively. 

```python
model.predict_proba(some_new_data)
model.predict_log_proba(some_new_data)
```

Measuring Performance: Cross-Validation
---------------------------------------

We never want to deploy a model without some understanding of how it would perform under realistic settings, getting some sense of a model's accuracy. For data scientists, the process of measuring model performance is usually called "evaluation". Evaluation is a topic with a long history, and much research. Fortunately scikit.learn makes the basics of evaluation easy.

Let's take the most simple case, where we have a model, some collection of labeled training data, and another collection of labeled data that we want to use for evaluation- compare the trained model's predictions to the actual target labels.

While the `score` method is a useful tool for evaluating models, it assumes you have a separate batch of labeled data dedicated for training. In practice, one is often struggling to get as much training data as possible; if you have this labeled data, you would want to use it to build even smarter models!

Fortunately, there is a tool for simultaniously using all available data for both training and evaluation, cross-validation. Cross-validation divides up the labeled data into k slices. Then, it iterates over these k, training a model on the remaining k-1 slices, using the current slice for evaluation. In the end, you have a better understanding of a model's perfromance after using all the data for both training and testing, and because you've made multiple estimates, you also have some understanding of the model's performance distribution.

To perform cross-validation, use the [`cross_val_score`](http://scikit-learn.org/stable/modules/generated/sklearn.cross_validation.cross_val_score.html) function in scikit's `cross_validation` package. This function takes 3 arguments, the model to be evaluated, a list of of lists of example features to be used, and a list of example labels.

As optional arguments, you can tell the cross-validator how many times to divide the data, using the `cv` argument, and you can tell it what performance measure to capture, using the `scoring` argument. Below are some of the possible metrics that are available for measuring a model's performance:

- accuracy: `sklearn.metrics.accuracy_score`: the percent of examples correctly classified.
- precision: `sklearn.metrics.precision_score`: considers the performance labeling a particular class, set by the additional argument `pos_label`. Measures the the percentage of examples labeled as belonging to that class that have been correctly labeled. 
- recall:	`sklearn.metrics.recall_score`: considers the performance labeling a particular class, set by the additional argument `pos_label`. Measures the the percentage of examples actually belonging to that class that have been correctly labeled.
- f1: `sklearn.metrics.f1_score`: An weighted average of precision and recall. As before, the class of interest can be set by an additional argument `pos_label`.

```python
import sklearn.cross_validation as cv
import sklearn.datasets as ds
import sklearn.svm as svm

digits = ds.load_digits()
classifier = svm.SVC(gamma=0.001, C = 100.0)
scores = cv.cross_val_score(classifier, digits.data, digits.target,
                            cv=20, scoring='accuracy')
print scores
print "Accuracy: %f ( + / - %f )" % (scores.mean(), scores.std() / 2)
```

Saving and retrieving a predictive model
----------------------------------------

Once we are happy with the performance of a predictive mdoel, we'll generally want to save it to a file so we can use it later. This file can be passed around, distributed, to different computers, and, when the time is right, loaded back into python to make predictions. 

Python provides a nice library for saving and loading models (or data structures generally), [`pickle`](https://wiki.python.org/moin/UsingPickle). 

```python
import pickle

# train a model
classifier = svm.SVC(gamma=0.001, C = 100.0)
classifier.fit(digits.data, digits.target)

# open a file for writing in binary mode
model_file = open('svc.file', 'wb')

# write the model to the file
pickle.dump(classifier, model_file)


model_file.flush()
model_file.close()


# read the model back in
# open a file for binary reading
model_file_r = open('svc.file', 'rb')
read_classifier = pickle.load(model_file_r)
model_file_r.close()

print read_classifier
```
